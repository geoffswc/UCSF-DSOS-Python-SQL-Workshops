```python

# ==========================================
# STEP 2: DOWNLOAD THE FILE FROM THE URL
# ==========================================
import os
import whisper
import torch

# 💡 REPLACE THIS WITH YOUR PUBLIC URL
public_url = "https://archive.org/download/tobacco_mxyh0015/tobacco_mxyh0015.mp4"

# Define a temporary local filename for the download
local_filename = "downloaded_audio.mp3"

print(f"\nDownloading audio from: {public_url}")
# Using curl to download the file into Colab's storage
os.system(f'curl -L "{public_url}" -o {local_filename}')

if os.path.exists(local_filename) and os.path.getsize(local_filename) > 0:
    print("✅ Download complete!")
else:
    raise FileNotFoundError("❌ Download failed. Please check the public URL.")

# ==========================================
# STEP 3: INITIALIZE WHISPER & TRANSCRIBE
# ==========================================
# Detect if a GPU is available, otherwise default to CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"\nLoading Whisper model on [{device.upper()}]...")

# Loading the 'base' model. It's a great balance of speed and accuracy for CPU.
# If it's too slow, try "tiny". If you want higher accuracy, try "small".
model = whisper.load_model("tiny", device=device)

print("Transcribing file... (this may take a few minutes on CPU)")
result = model.transcribe(local_filename)

# Store the final transcript text in a variable
raw_transcript = result["text"]
```
