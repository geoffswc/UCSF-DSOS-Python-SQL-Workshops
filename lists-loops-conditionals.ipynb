```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
temperatures = [71, 64, 57, 78, 72, 58, 60]
humidities = [.5, .6, .7, .8, .5, .6, .7]
```

```python
for i in range(len(days)):
  weather_index = temperatures[i] + humidities[i] * 10
  print(days[i], temperatures[i], humidities[i], weather_index)
```

```python
# what if I wanted to know which days meet the conditions for a certain metric?
# only for temperature above a threshold, or a humidity, or some calculation involvoing both?
```

```python
# print the temperature only when it is above 65
```

```python
for i in range(len(days)):
  print(i)
  if temperatures[i] > 70:
    print("warm", temperatures[i])
  elif temperatures[i] > 66:
    print("medium", temperatures[i])
  elif temperatures[i] > 60:
    print("cool", temperatures[i])
```
