```python
# prompt: create bar chart extracting italy to poland and use a prettier color palate

import pandas as pd
import matplotlib.pyplot as plt

# Assuming df_subset is already defined as in the previous code
# If not, recreate it here:
df_europe = pd.read_csv('https://raw.githubusercontent.com/geoffswc/IntroToPythonPart2/refs/heads/master/data/gapminder_gdp_europe.csv', index_col='country')
df_subset = df_europe.loc['Italy':'Poland', 'gdpPercap_1972':'gdpPercap_1992']


df_subset.T.plot(kind='bar', colormap='viridis') # Using the 'viridis' colormap
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.xlabel("Year")
plt.ylabel("GDP per capita")
plt.title("GDP per capita from 1972 to 1992 (Italy to Poland)")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
